```python
import json, io, time,os
import requests
import pandas as pd
import alpha_vantage

pd.options.display.max_rows = 5000
```

```python
def get_data(symbol):
    API_URL = "https://www.alphavantage.co/query"
    data = {
      "function": "TIME_SERIES_DAILY",
      "symbol": "NSE:"+symbol,
      "datatype": "csv",
      "apikey": '46N9W9ZQM2YTWC1J',
      }
    response = requests.get(API_URL, data)
    response_content=response.content
    required_data = pd.read_csv(io.StringIO(response_content.decode('utf-8')))
    return required_data
```

```python
get_data('KOTAKBANK')
```

```python
def save_csv_file(symbol):
    try:
        data=get_data(symbol)
        #display(data)
        data.to_csv('/home/ronith/Education/FinalYearProject/StockPrices/'+symbol+'.csv')
        print(symbol,' Saved Successfully')
    except Exception as Error:
        print(symbol,Error)
```
